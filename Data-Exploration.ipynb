```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DATA/train.csv")
```

# Investigate each column

```python
# Brand
brand_counts = df.Brand.value_counts(dropna=False)
brand_counts.plot(kind='bar')
plt.title("Brand");
plt.tight_layout()
plt.savefig('brand_bar.png')
```

```python
# Material
material_counts = df.Material.value_counts(dropna=False)
material_counts.plot(kind='bar')
plt.title("Material");
plt.tight_layout()
plt.savefig('material_bar.png')
```

```python
# Material
size_counts = df.Size.value_counts(dropna=False)
size_counts.plot(kind='bar')
plt.title("Sizes");
plt.tight_layout()
plt.savefig('sizes_bar.png')
```

```python
# Compartments
print(df.Compartments.isna().sum())
df.Compartments.hist(bins=10)
plt.title("Compartments");
plt.tight_layout()
plt.savefig('compartments_hist.png')
```

```python
# Laptop Compartment
laptop_counts = df['Laptop Compartment'].value_counts(dropna=False)
laptop_counts.plot(kind='bar');
plt.tight_layout()
plt.savefig('laptop_compartment_bar.png')
```

```python
# Waterproof
waterproof_counts = df.Waterproof.value_counts(dropna=False)
waterproof_counts.plot(kind='bar');
plt.tight_layout()
plt.savefig('waterproof_bar.png')
```

```python
# Style
style_counts = df.Style.value_counts(dropna=False)
style_counts.plot(kind='bar')
plt.title('Style');
plt.tight_layout()
plt.savefig('style_bar.png')
```

```python
# Color
color_counts = df.Color.value_counts(dropna=False)
color_counts.plot(kind='bar')
plt.title('Color');
plt.tight_layout()
plt.savefig('color_bar.png')
```

```python
# Weight Capacity (kg)
print(df['Weight Capacity (kg)'].isna().sum())
df['Weight Capacity (kg)'].hist(bins=100)
plt.title('Weight Capacity (kg) Distribution');
plt.tight_layout()
plt.savefig('weight_hist.png')
```

```python
df.Price.hist()
plt.title('Price Distribution')
plt.tight_layout()
plt.savefig('price_hist.png')
```
